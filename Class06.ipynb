# **Incident Response**
Information security incident response is a structured process that helps organizations detect, contain, analyze, and recover from security incidents. It begins with early identification through monitoring and detection techniques, followed by rapid containment to limit damage. Proper data collection and preservation are essential, including forensic copies and thorough documentation to maintain evidence integrity. Data analysis and visualization help uncover the incident’s cause, scope, and impact. Maintaining a strict chain of custody and following standards like ISO 27037 ensure that digital evidence remains reliable and legally admissible.

```python
# Importing all the necessary libraries and resources:
import matplotlib.pyplot as plt
```

## **Log Analysis for Incident Detection**
The example below example that simulates log analysis to identify suspicious activity, a key step in incident response.

```python
# Sample log entries:
logs = [
    '2025-12-10 08:15:23 INFO User login successful',
    '2025-12-10 08:17:45 WARNING Failed login attempt',
    '2025-12-10 08:18:02 WARNING Failed login attempt',
    '2025-12-10 08:18:30 WARNING Failed login attempt',
    '2025-12-10 08:19:10 INFO User logout'
]

# Detecting suspicious activity (multiple failed logins in this case):
failed_attempts = [log for log in logs if 'Failed login' in log]

# Threshold-based alert:
if len(failed_attempts) >= 3:
    print('⚠️ Suspicious activity detected: Multiple failed login attempts')
    for attempt in failed_attempts:
        print(attempt)
else:
    print('No suspicious activity detected.')
```

## **Visualizing Network Activity Spikes**
The example below demonstrates how visualization can help identify anomalies during an incident response by highlighting unusual network traffic patterns.

```python
# Simulated network traffic data; in this case requests per hour:
hours = list(range(24))
network_requests = [
    120, 110, 105, 100, 98, 95, # Night (normal)
    130, 150, 170, 200, 220, 240, # Morning
    260, 280, 300, 650, 700, 680, # Afternoon (suspicious spike)
    320, 280, 260, 230, 200, 180 # Evening
]

# Plotting traffic data:
plt.figure(figsize=(10, 4))
plt.plot(hours, network_requests, marker='o')
plt.title('Network Traffic Analysis')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Requests')
plt.grid(True)

# Highlighting suspicious threshold:
threshold = 500
plt.axhline(y=threshold, color='r', linestyle='--', label='Suspicious Threshold')
plt.legend()

plt.show()
```
